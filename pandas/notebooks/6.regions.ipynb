```python
import pandas as pd
import json
```

```python
spending = pd.read_csv('../output/spending.csv')

f = open('../../local-area.json', "r")
regions = json.loads(f.read())
```

```python
for r in regions['features']:
    name = r['properties']['LAD13NM']
    region = spending.loc[lambda spending: spending['brexit_constituency'].str.upper() == name.upper()].iloc[0].region
    r['properties']['region'] = region
    print(name)
    print(region)
```

```python
with open('../../local-area-regions.json', 'w') as f:
    json.dump(regions, f)
```
