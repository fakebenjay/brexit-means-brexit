```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
spending = pd.read_csv('../output/spending-postcode.csv')
```

```python
spending = spending.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
dist_spending = spending.groupby(['brexit_constituency']).sum()
```

```python
vote_results = pd.read_csv('../data/vote-results-with-ni.csv')
```

```python
result_areas = list(vote_results['Area'])
```

```python
dist_spending.insert(0, 'Area_Code', 'BLANK')
dist_spending.insert(1, 'Region', 'BLANK')
dist_spending.insert(2, 'Region_Code', 'BLANK')
dist_spending['Electorate'] = 0
dist_spending['Votes_Cast'] = 0
dist_spending['Valid_Votes'] = 0
dist_spending['Pct_Turnout'] = 0
dist_spending['Remain'] = 0
dist_spending['Leave'] = 0
dist_spending['Pct_Remain'] = 0
dist_spending['Pct_Leave'] = 0
dist_spending['Rejected_Ballots'] = 0
dist_spending['Pct_Rejected'] = 0
```

```python
dist_spending
```

```python
def add_col(col_name, df, ref):
    counter = 0
    for council in list(df.index):
        counter += 1
        value = ref.loc[ref['Area'] == council][col_name].values[0]
        df.loc[council, col_name] = value
        print(str(counter) + '. ' + council + ', ' + str(value))
```

```python
add_col('Electorate', dist_spending, vote_results)
print('-------------')
add_col('Votes_Cast', dist_spending, vote_results)
print('-------------')
add_col('Valid_Votes', dist_spending, vote_results)
print('-------------')
add_col('Pct_Turnout', dist_spending, vote_results)
print('-------------')
add_col('Remain', dist_spending, vote_results)
print('-------------')
add_col('Leave', dist_spending, vote_results)
print('-------------')
add_col('Pct_Remain', dist_spending, vote_results)
print('-------------')
add_col('Pct_Leave', dist_spending, vote_results)
print('-------------')
add_col('Rejected_Ballots', dist_spending, vote_results)
print('-------------')
add_col('Pct_Rejected', dist_spending, vote_results)
print('-------------')
```

```python
dist_spending.insert(6, 'spending_per_voter', 0)
```

```python
dist_spending.columns = map(str.lower, dist_spending.columns)
```

```python
dist_spending = dist_spending.assign(spending_per_voter = dist_spending['project_eu_contribution_gbp'] / dist_spending['electorate'])
```
