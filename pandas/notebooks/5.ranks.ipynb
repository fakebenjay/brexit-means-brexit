```python
import pandas as pd
```

```python
spending = pd.read_csv('../output/spending.csv')
```

```python
spending = spending.sort_values(by=['pct_remain'], ascending=False)
counter = 1

for index, row in spending.iterrows():
    spending.loc[index, 'remain_rank'] = int(counter)
    counter += 1
```
