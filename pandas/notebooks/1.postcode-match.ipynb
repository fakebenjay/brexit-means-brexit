```python
import pandas as pd
```

```python
postcodes = pd.read_csv('../data/National_Statistics_Postcode_Lookup_UK.csv', low_memory=False)
```

```python
postcodes['Local Authority Name']
```

```python
spending = pd.read_csv('../data/spending.csv')
```

```python
def gbp_to_float(string):
    if isinstance(string, str):
        string = string.replace('', '').replace(',', '')
        return float(string)
    else:
        return 0.00
```

```python
def round_to_two(num):
    return round(num, 2)
```

```python
spending.project_eu_contribution_gbp = spending.project_eu_contribution_gbp.apply(gbp_to_float)
```

```python
spending.eu_contribution_gbp = spending.eu_contribution_gbp.apply(gbp_to_float)
```

```python
spending.project_total_cost_gbp = spending.project_total_cost_gbp.apply(gbp_to_float)
```

```python
spending_2 = pd.DataFrame({
    'eu_contribution_gbp': spending.eu_contribution_gbp,
    'project_eu_contribution_gbp': spending.project_eu_contribution_gbp,
    'project_total_cost_gbp': spending.project_total_cost_gbp,
    'postcode': spending.postcode,
    'parliamentary_constituency_name': spending.parliamentary_constituency_name
})
```

```python
ni = ['Belfast East','Belfast North','Belfast South','Belfast West','East Antrim','East Londonderry','Fermanagh and South Tyrone','Foyle','Lagan Valley','Mid Ulster','Newry and Armagh','North Antrim','North Down','South Antrim','South Down','Strangford','Upper Bann','West Tyrone']
```

```python
postcodes.loc[postcodes['Postcode 2'] == 'AB12 4NB']
```

```python
spending_2['brexit_constituency'] = ''
```

```python
for index, row in spending_2.iterrows():
#     if index < 11903: continue
    
    if spending_2.loc[index, 'parliamentary_constituency_name'] in ni:
        dist = spending_2.loc[index, 'parliamentary_constituency_name']
        spending_2.loc[index, 'brexit_constituency'] = dist
        print(str(index) + '. ' + str(dist))
    else:
        dist_row = postcodes.loc[postcodes['Postcode 3'] == spending_2.loc[index, 'postcode']]
        if len(dist_row) == 0:
            dist = "OOPSY I GOT A STICKY WICKET"
        else:
            dist = dist_row['Local Authority Name'].values[0]
            
        spending_2.loc[index, 'brexit_constituency'] = dist
        print(str(index) + '. ' + str(dist))
```

```python
spending_2.to_csv('../output/spending-postcode.csv')
```
