```python
import pandas as pd
import numpy as np
```

```python
spending = pd.read_csv('../output/dist-spending.csv')
```

```python
population = pd.read_csv('../data/dist-population.csv')
```

```python
population = population.drop(columns=['Constituency type', 'Total electors 2016', 'Attainers 2015', 'Attainers 2016', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'])
```

```python
population['eu_spending'] = 0
```

```python
spending.loc[spending['parliamentary_constituency_name'] == 'Sedgefield'].project_eu_contribution_gbp
```

```python
spending.loc[spending['parliamentary_constituency_name'] == 'Sedgefield'].project_eu_contribution_gbp.values[0]
```

```python
not spending.loc[spending['parliamentary_constituency_name'] == 'Wales'].empty
```

```python
for index, row in population.iterrows():
    if not spending.loc[spending['parliamentary_constituency_name'] == row['Parliamentary constituency']].empty:
        population.loc[index, 'eu_spending'] = spending.loc[spending['parliamentary_constituency_name'] == row['Parliamentary constituency']]['project_eu_contribution_gbp'].values[0]
```

```python
population
```

```python
population = population.rename(index=str, columns={"Code": "code", "Parliamentary constituency": "constituency", "Total electors 2015": "tot_electors_2015"})
```

```python
population['spending_per_elector'] = round(population['eu_spending']/population['tot_electors_2015'], 2)
```

```python
population.to_csv('../output/pop-and-spending.csv')
```
